```python
import pandas as pd
import numpy as np
```

Import raw data.
Clean up blank lines and lines containing descriptions, legend, etc.

```python
data2021 = pd.read_csv("raw_data/hcny_all_tables_2021.csv", skiprows=4)
data2020 = pd.read_csv("raw_data/hcny_all_tables_2020.csv", skiprows=4)
data2019 = pd.read_csv("raw_data/hcny_all_tables_2019.csv", skiprows=4)
data2018 = pd.read_csv("raw_data/hcny_all_tables_2018.csv", skiprows=4)
data2017 = pd.read_csv("raw_data/hcny_all_tables_2017.csv", skiprows=4)
data2016 = pd.read_csv("raw_data/hcny_all_tables_2016.csv", skiprows=4)

data2021 = data2021[data2021['State'].str.contains('Represents zero') == False]
data2021 = data2021[data2021['State'].str.contains('Less than half') == False]
data2021 = data2021[data2021['State'].str.contains('Tracheal mites') == False]
data2021 = data2021[data2021['State'].str.contains('Includes American') == False]
data2021 = data2021[data2021['State'].str.contains('Includes weather') == False]
data2021 = data2021[data2021['State'].str.contains('Includes data') == False]
data2021 = data2021[data2021['State'].str.contains('Honey Bee Colonies') == False]
data2021 = data2021[data2021['State'].str.contains('Colony Health') == False]
data2021 = data2021[data2021['State'].str.contains('Percent of colonies') == False]
data2021 = data2021[data2021['State'].str.contains('Not applicable') == False]
data2021 = data2021[data2021['State'].str.contains('plus all colonies') == False]
data2021 = data2021[data2021['State'].str.contains('Percent lost') == False]
data2021 = data2021[data2021['State'].str.contains('Defined as any') == False]
data2021 = data2021[data2021['State'].str.contains('Percent renovated') == False]
data2021 = data2021[data2021['State'].str.contains('Colonies, Maximum') == False]
data2021 = data2021[data2021['State'].str.contains('Colonies Lost') == False]
data2021 = data2021[data2021['State'].str.contains('Loss reported') == False]
```

Drop superfluous index columns.

Get rid of rows with a NaN state.

Set State column as index.

```python
data2021.drop(columns=['5', 'h'], axis=1, inplace=True)
data2021 = data2021[data2021['State'].notna()]
data2021.set_index('State')
```

It turns out that each csv file has the full set of the previous year's data and then the first two quarters of the year in the file name. Assuming that the most recent version is the most recently collected and has the most up to date data, we'll keep the most recently collected version.

So, deleting the rows that are the first two quarters of the incomplete year of data, except for the 2021 datasheet, since that will have all the 2020 data, and the first two quarters (all that exists) of the 2021 data. 

The data has to be filled in chunks since the first six columns for all the states are listed for each quarter for 2020, then the first six columns for 2021 in the next set of rows, then the last six columns for all states for 2021, then the last six for 2021. 

```python
data2021.insert(1, 'year', np.nan)
last_year_values = {'year': '2020'}
this_year_values = {'year': '2021'}
data2021.fillna(value=last_year_values, limit=191, inplace=True)
data2021.fillna(value=this_year_values, limit=96, inplace=True)
data2021.fillna(value=last_year_values, limit=192, inplace=True)
data2021.fillna(value=this_year_values, limit=96, inplace=True)

data2021.insert(2, 'quarter', np.nan)
values_jan = {'quarter': 'Jan-Mar'}
values_apr = {'quarter': 'Apr-Jun'}
values_jul = {'quarter': 'Jul-Sep'}
values_oct = {'quarter': 'Oct-Dec'}
data2021.fillna(value=values_jan, limit=47, inplace=True)
data2021.fillna(value=values_apr, limit=48, inplace=True)
data2021.fillna(value=values_jul, limit=48, inplace=True)
data2021.fillna(value=values_oct, limit=48, inplace=True)
data2021.fillna(value=values_jan, limit=48, inplace=True)
data2021.fillna(value=values_apr, limit=48, inplace=True)

data2021.fillna(value=values_jan, limit=48, inplace=True)
data2021.fillna(value=values_apr, limit=48, inplace=True)
data2021.fillna(value=values_jul, limit=48, inplace=True)
data2021.fillna(value=values_oct, limit=48, inplace=True)
data2021.fillna(value=values_jan, limit=48, inplace=True)
data2021.fillna(value=values_apr, limit=48, inplace=True)

# Set multi-index, partly because doing all the above somehow restored the default numeric index
data2021.set_index(['State', 'year', 'quarter'])
```

The first six columns for each state are stored in different rows from the second six columns for a state. 
Cut the dataframe apart and join those so that they're all in one row.

```python
bottom_half = data2021.iloc[288:]
data2021 = data2021.iloc[:287]
bottom_half.set_index(['State', 'year', 'quarter'])
data2021.set_index(['State', 'year', 'quarter'])

print(data2021.shape)
print(bottom_half.shape)

# This is where it messes up: What I want is a Dataframe that has a shape of (290, 20) 
# with the information from the 'bottom_half' dataframe is attached as columns 11-20 

# I also don't know why that numeric index keeps coming back when I've set the index to be the state column
# I assumed that it had to the State column to be able to join the two together

# When I try concat, it makes a dataframe that has a shape of (580, 10), attaching the 'bottom_half' dataframe
# after the first one rather than expanding the columns
# When I try join, it gets the columns, but they're all NaN 

# merged_data2021 = pd.concat([data2021, bottom_half], axis=1)
merged_data2021 = data2021.join(bottom_half, lsuffix='_left', rsuffix='_right')
print(merged_data2021)
merged_data2021.to_csv('merged2021.csv')
print(merged_data2021.shape)
```
