The first seven columns for each state are stored in different rows (the upper half of each csv file) from the second seven columns for each state (the lower half of each csv file). 
Cut the dataframe apart and specify the indices for both.
Set the column names for each now that they're separate, since the columns change names halfway through the larger sheet. Lastly, with the dataframes still split, remove the empty column from "bottom half" and the columns on renovated colonies, since they won't be used.

```python
bottom_half_2016 = data2016.iloc[240:482]
top_half_2016 = data2016.iloc[:239]

bottom_half_2017 = data2017.iloc[288:578]
top_half_2017 = data2017.iloc[:287]

bottom_half_2018 = data2018.iloc[288:578]
top_half_2018 = data2018.iloc[:287]

bottom_half_2019 = data2019.iloc[288:]
top_half_2019 = data2019.iloc[:287]

bottom_half_2020 = data2020.iloc[288:]
top_half_2020 = data2020.iloc[:287]

bottom_half_2021 = data2021.iloc[288:]
top_half_2021 = data2021.iloc[:287]

bottom_half_2016.set_index(['State', 'year', 'quarter'])
top_half_2016.set_index(['State', 'year', 'quarter'])

bottom_half_2017.set_index(['State', 'year', 'quarter'])
top_half_2017.set_index(['State', 'year', 'quarter'])

bottom_half_2018.set_index(['State', 'year', 'quarter'])
top_half_2018.set_index(['State', 'year', 'quarter'])

bottom_half_2019.set_index(['State', 'year', 'quarter'])
top_half_2019.set_index(['State', 'year', 'quarter'])

bottom_half_2020.set_index(['State', 'year', 'quarter'])
top_half_2020.set_index(['State', 'year', 'quarter'])

bottom_half_2021.set_index(['State', 'year', 'quarter'])
top_half_2021.set_index(['State', 'year', 'quarter'])

first_columns = {'January 1': 'Starting colonies', 'Unnamed: 4': 'Maximum colonies', 'Unnamed: 5': 'Lost colonies', 'Unnamed: 6': 'Percent lost', 'Unnamed: 7': 'Added colonies', 'Unnamed: 8': 'Renovated colonies', 'Unnamed: 9': 'Percent renovated'}
second_columns = {'January 1': 'Varroa mites', 'Unnamed: 4': 'Other pests', 'Unnamed: 5': 'Diseases', 'Unnamed: 6': 'Pesticides', 'Unnamed: 7': 'Other', 'Unnamed: 8': 'Unknown', 'Unnamed: 9': 'empty'}
alt_first_columns = {'Unnamed: 3': 'Starting colonies', 'Unnamed: 4': 'Maximum colonies', 'Unnamed: 5': 'Lost colonies', 'Unnamed: 6': 'Percent lost', 'Unnamed: 7': 'Added colonies', 'Unnamed: 8': 'Renovated colonies', 'Unnamed: 9': 'Percent renovated'}
alt_second_columns = {'Unnamed: 3': 'Varroa mites', 'Unnamed: 4': 'Other pests', 'Unnamed: 5': 'Diseases', 'Unnamed: 6': 'Pesticides', 'Unnamed: 7': 'Other', 'Unnamed: 8': 'Unknown', 'Unnamed: 9': 'empty'}

top_half_2016.rename(alt_first_columns, axis=1, inplace=True)
bottom_half_2016.rename(alt_second_columns, axis=1, inplace=True)

top_half_2017.rename(alt_first_columns, axis=1, inplace=True)
bottom_half_2017.rename(alt_second_columns, axis=1, inplace=True)

top_half_2018.rename(alt_first_columns, axis=1, inplace=True)
bottom_half_2018.rename(alt_second_columns, axis=1, inplace=True)

top_half_2019.rename(first_columns, axis=1, inplace=True)
bottom_half_2019.rename(second_columns, axis=1, inplace=True)

top_half_2020.rename(first_columns, axis=1, inplace=True)
bottom_half_2020.rename(second_columns, axis=1, inplace=True)

top_half_2021.rename(first_columns, axis=1, inplace=True)
bottom_half_2021.rename(second_columns, axis=1, inplace=True)

top_half_2016.drop(['Renovated colonies', 'Percent renovated'], inplace=True, axis=1)
bottom_half_2016.drop('empty', inplace=True, axis=1)

top_half_2017.drop(['Renovated colonies', 'Percent renovated'], inplace=True, axis=1)
bottom_half_2017.drop('empty', inplace=True, axis=1)

top_half_2018.drop(['Renovated colonies', 'Percent renovated'], inplace=True, axis=1)
bottom_half_2018.drop('empty', inplace=True, axis=1)

top_half_2019.drop(['Renovated colonies', 'Percent renovated'], inplace=True, axis=1)
bottom_half_2019.drop('empty', inplace=True, axis=1)

top_half_2020.drop(['Renovated colonies', 'Percent renovated'], inplace=True, axis=1)
bottom_half_2020.drop('empty', inplace=True, axis=1)

top_half_2021.drop(['Renovated colonies', 'Percent renovated'], inplace=True, axis=1)
bottom_half_2021.drop('empty', inplace=True, axis=1)
```

Now that the two datasets have indices and correct column labels, join them together in one set. 

```python
merged_data2021 = pd.merge_ordered(top_half_2021, bottom_half_2021, on=['State', 'year', 'quarter'])
merged_data2020 = pd.merge_ordered(top_half_2020, bottom_half_2020, on=['State', 'year', 'quarter'])
merged_data2019 = pd.merge_ordered(top_half_2019, bottom_half_2019, on=['State', 'year', 'quarter'])
merged_data2018 = pd.merge_ordered(top_half_2018, bottom_half_2018, on=['State', 'year', 'quarter'])
merged_data2017 = pd.merge_ordered(top_half_2017, bottom_half_2017, on=['State', 'year', 'quarter'])
merged_data2016 = pd.merge_ordered(top_half_2016, bottom_half_2016, on=['State', 'year', 'quarter'])
```

Clean up the last few junk rows in each sheet. 

```python
merged_data2016 = merged_data2016[merged_data2016['State'].str.contains('State|Year|Quarter|Item|2016') == False]
merged_data2017 = merged_data2017[merged_data2017['State'].str.contains('State|Year|Quarter|Item|2016|2017') == False]
merged_data2018 = merged_data2018[merged_data2018['State'].str.contains('State|Year|Quarter|Item|2017|2018') == False]
merged_data2019 = merged_data2019[merged_data2019['State'].str.contains('State|Year|Quarter|Item|2018|2019') == False]
merged_data2020 = merged_data2020[merged_data2020['State'].str.contains('State|Year|Quarter|Item|2019|2020') == False]
merged_data2021 = merged_data2021[merged_data2021['State'].str.contains('State|Year|Quarter|Item|2020|2021') == False]

merged_2016_2017 = pd.concat([merged_data2016, merged_data2017])
merged_2016_2018 = pd.concat([merged_2016_2017, merged_data2018])
merged_2016_2019 = pd.concat([merged_2016_2018, merged_data2019])
merged_2016_2020 = pd.concat([merged_2016_2019, merged_data2020])
all_merged = pd.concat([merged_2016_2020, merged_data2021])

all_merged.to_csv('all_merged.csv')
```
